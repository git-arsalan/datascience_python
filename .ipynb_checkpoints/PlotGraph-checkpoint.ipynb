# Adding Titles

```python
#import libaraires
import seaborn as sns
import matplotlib.pyplot as plt

# import datasets
phool = sns.load_dataset("iris")
type(phool)

phool

# draw a lineplot
# line plot works on two numerical value
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

# Adding Limits

```python
#import libaraires
import seaborn as sns
import matplotlib.pyplot as plt

# import datasets
phool = sns.load_dataset("iris")
type(phool)

phool

# draw a lineplot
# line plot works on two numerical value
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set Styles

```python
#import libaraires
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(style=None, rc=None)

# import datasets
phool = sns.load_dataset("iris")
type(phool)

phool

# draw a lineplot
# line plot works on two numerical value
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)
sns.set_style("dark")
plt.show()
```

# Size of Figure

```python
#import libaraires
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(style=None, rc=None)

# import datasets
phool = sns.load_dataset("iris")
#type(phool)

#phool

# draw a lineplot
# line plot works on two numerical value
plt.figure(figsize=(4,2))
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)

plt.title("Phoolon ka Plot")
#plt.xlim(2)
#plt.ylim(1)
sns.set_style("dark")
plt.show()
```
